```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
```

Finds the element by the live-table id which contains the whole data in it

```python
table = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.ID, 'live-table')))
```

Beautiful Soup

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(table.get_attribute('innerHTML'), 'lxml')
```

Each match data is inside a dict with this class event_round(i.e. final,SF,QF,...etc)

```python
match_types = soup.find_all("div", {"class": "event__round event__round--static"})
```

```python
rounds = dict()
```

Since every match type has its sibling i.e. up until the next match type. For eg: the semi final comes after the final , then the match type class is event_round and this is where we have to stop , since say in the semi-finals there are 2 matches , we need to get the details of the matches upto the semi finals and then again loop to quarter finals and get the matches under the quarter finals and so and so forth. Also we need to store it in a dictionary.

```python
'''Loop through all the match types(eg: final,ssemifinal,quarterfinal)'''
for x in match_types:
        rounds[x.get_text()] = list()
        '''looks at div next to the finals tab and adds it to the final match type and also looks for the same for semi final,
        since each match under the finals div is a sibling to the finals div , similarily each match under the semi finals div
        is a sibling to the semi finals div and so on and so forth'''
        for y in x.next_siblings:
               if y['class'][0] == 'event__round' or y['class'][0] == 'event__more':
                    break             
               rounds[x.get_text()].append(y)
                
```

Now since the dictionary created above has only the generator elements and we need to granulate further and move down to further granular details up until the modt lowest level , we need to create a dictionary with further details

```python
rounds_granular = dict()
```

Every round has a match (for eg: - the final has 1 match , the semi final has 2 matches , the quaarter final has 4 matcches and so on and so forth , so the details of every round needs to be properly stored and created.

```python
import pandas as pd
```

```python
for rounds_,matches_ in rounds.items():
    '''Crete a list for each round , since each round can have more than 1 match'''
    rounds_granular[rounds_] = list()
    '''Add the details for each match in a dict'''
    for match in matches_:
        x = dict()
        '''The date is inside a div with the class event time and has the following format and hence needs to be converted
        into that corresponding format for better datetime conversion'''
        x['date'] = pd.to_datetime(match.find("div",{"class":"event__time"}).get_text()+".2022",format="%d.%m. %H:%M.%Y")
        '''The winner of the match comes under the div with the class event participant home'''
        x['W'] = match.find("div",{"class":"event__participant--home"}).get_text()
        '''The country of the  winner of the match comes under the div with the class event logo home'''
        x['W_Country'] = match.find("span",{"class":"event__logo--home"})['title']
        '''The loser of the match comes under the div with the class event participant away'''
        x['L'] = match.find("div",{"class":"event__participant--away"}).get_text()
        '''The country of the  loser of the match comes under the div with the class event participant away'''
        x['L_Country'] = match.find("span",{"class":"event__logo--away"})['title']
        '''mws stands for match score of winner'''
        x['mws'] = match.find("div",{"class":"event__score--home"}).get_text()
        '''mls stands for match score for loser'''
        x['mls'] = match.find("div",{"class":"event__score--away"}).get_text()
        '''since it is a 5 set macth we loop through all the 5 numbers in the set'''
        for i in range(5):
            '''Every set score is inside this classs'''
            y = match.select("div.event__part--home.event__part--"+str(i+1))
            z = match.select("div.event__part--away.event__part--"+str(i+1))
            '''We check for the set score since it might so happen that the match did not go to all the 5 sets and
            was ended before that and hence it is checked if it goes the distance'''
            if y and z:
                '''also some fields have the subscript mentioned in them if it goes to a tie breaker and we check if the length
                of the game is more than 2 , if it is , it means that it needs to be reduced to one , since a roland garros
                grand slam set has only single digit games mentioned in them and these need to be accounted for'''
                x['s'+str((i+1))+'w'] = y[0].get_text() if(len(y[0].get_text()) == 1) else y[0].get_text()[0]
                x['s'+str((i+1))+'l'] = z[0].get_text() if(len(z[0].get_text()) == 1) else z[0].get_text()[0]
            else:
                '''if it doesnt go the distance we simply put 0 in the set score for those unplayed sets'''
                x['s'+str((i+1))+'w'] = 0
                x['s'+str((i+1))+'l'] = 0
        rounds_granular[rounds_].append(x)
                
        
        
```

```python
column_names = ['round','date','p_name','p_country','pm_score','s1_score','s2_score','s3_score','s4_score','s5_score','result']
```

```python
match_df = pd.DataFrame(columns = column_names)
```

```python
match_df
```

```python
'''every dictionary value has both the winner and the loser details so simply seperate them out into 2 different 
rows of the datframe to carry out much easier analysis of the data , this makes it so much better to do a lot of analysis'''
for mtype,value in rounds_granular.items():
    for match in value:
        match_df = match_df.append({'round':mtype,'date':match['date'],'p_name':match['W'],'p_country':match['W_Country'],
                         'pm_score':int(match['mws']),'s1_score':int(match['s1w']),'s2_score':int(match['s2w']),
                          's3_score':int(match['s3w']),'s4_score':int(match['s4w']),'s5_score':int(match['s5w']),
                           'result':'W' if int(match['mws']) > int(match['mls']) else 'L'},ignore_index=True)
        
        match_df = match_df.append({'round':mtype,'date':match['date'],'p_name':match['L'],'p_country':match['L_Country'],
                         'pm_score':int(match['mls']),'s1_score':int(match['s1l']),'s2_score':int(match['s2l']),
                          's3_score':int(match['s3l']),'s4_score':int(match['s4l']),'s5_score':int(match['s5l']),
                           'result':'W' if int(match['mls']) > int(match['mws']) else 'L'},ignore_index=True)
        
        
        
        
        
        
```

### Distribution of set of matches(i.e. no of 1,2(retd. hurt),3 set matches , 4 set matches and 5 set matches) by each player

Since the data has been extracted in a manner that every second player is matched up against the player above him in that round,
it makes sense to shift the data and take the opponent's players details along the same row , now if we add the number of sets for each pm_score, it will give the total sets played by both players in each round and hence we can create a set count only from every 2nd row of the group , also we could also sort by date and get similar results , but since it is already in order , the requirement to do so was not needed. The number of sets does not exceed 5 from the distribution. It can be checked by changing the nbins , it still remains 5 also. This analysis in particular required a lot of digging deep and has helped in merging all the lessons taught in the class.

```python
match_dist = match_df
```

```python
match_dist['player_2'] = match_dist['p_name'].shift()
```

```python
match_dist['pm_x'] = match_dist['pm_score'].shift()
```

```python
match_dist = match_dist[match_dist.index%2 != 0]
```

```python
match_dist['sets_played'] = match_dist['pm_score']+match_dist['pm_x']
```

```python
players = match_df.p_name.unique()
```

```python
players = np.append(players,'Total')
```

```python
def create_player_histo(player):
    if player != 'Total':
        player_dist = match_dist[(match_dist['p_name'] == player) | (match_dist['player_2'] == player)]
    else:
        player_dist = match_dist
    fig = px.histogram(player_dist,x='sets_played',nbins = 5)
    fig.show()
```

```python
widgets.interact(create_player_histo, player=players);
```
